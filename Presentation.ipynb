This notebook contains the code I used to generate scatterplots for the presentation.  
The images can be found in the 'images' folder.

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

# View Scatterplot of Raw Data

```python

# Load the data
data = pd.read_csv('data/aircraft-data_nov_dec.csv')
data = data[['alt', 'mph']]

# Plot the data to show the odd data points
plt.figure(figsize=(10, 6))
plt.scatter(data['alt'], data['mph'])
plt.xlabel('Altitude (ft)')
plt.ylabel('Speed (mph)')
plt.title('Aircraft Data')
plt.savefig('images/plot_raw.png')
plt.show()
```

# Preprocessing Steps
1) Remove outliers
2) Remove line of odd values

```python
# Remove the odd data points
data_no_fill = data[data['mph'] != 460]

# Plot the data to show the odd data points
plt.figure(figsize=(10, 6))
plt.scatter(data_no_fill['alt'], data_no_fill['mph'])
plt.xlabel('Altitude (ft)')
plt.ylabel('Speed (mph)')
plt.title('Aircraft Data')
plt.savefig('images/plot_cleaned_with_outliers.png')
plt.show()
```

```python
# Remove outliers (outer 2%)
data_no_fill = data_no_fill[(data_no_fill['alt'] > data_no_fill['alt'].quantile(0.01)) & (data_no_fill['alt'] < data_no_fill['alt'].quantile(0.99))]
data_no_fill = data_no_fill[(data_no_fill['mph'] > data_no_fill['mph'].quantile(0.01)) & (data_no_fill['mph'] < data_no_fill['mph'].quantile(0.99))]

# Plot the data to show the odd data points
plt.figure(figsize=(10, 6))
plt.scatter(data_no_fill['alt'], data_no_fill['mph'])
plt.xlabel('Altitude (ft)')
plt.ylabel('Speed (mph)')
plt.title('Aircraft Data')
plt.savefig('images/plot_cleaned_no_outliers.png')
plt.show()
```
